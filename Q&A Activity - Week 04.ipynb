```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('Telco-Customer-Churn.xlsx.csv')
df.head()
```

```python
#values on columns
df.columns.values
```

```python
#datatypes present in the dataset
df.dtypes
```

### 1. Compare churn count with respect to gender.

```python
sns.countplot(x ='gender', hue='Churn', data=df)
plt.title("gender vs churn count")
plt.show()
```

### 2. Find out how many female senior citizens there in the dataset  

```python
fs=df[df["gender"]=="Female"]['SeniorCitizen']==1
fs.value_counts()
```

this shows there are 568 female senior citzens in list

### 3.Compare 'tenure' with 'Total Charges' 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["tenure"], y=df["TotalCharges"])
plt.xlabel('Tenure (Months)')
plt.title("Tenure vs Total Charges")
plt.tight_layout()
```

###  4.Find out which contract is preferred by the senior citizen. 

```python
sc=df[df['SeniorCitizen']==1.0]
sns.countplot(x ='Contract', data=sc, color='lightgreen')
sns.set(rc={'axes.facecolor':'black', 'figure.facecolor':'white'})
plt.title("Contracts preferred by Senior Citizens")
plt.show()
```

From the above derived graph ,month to month contracts are preferred by senior citizens

### 5.Comment on your finds on Payment Method? 

```python
df['PaymentMethod'].value_counts()
```

```python
labels=['Electronic Check','Mailed Check','Bank Transfer','Credit Card']
plt.pie(df['PaymentMethod'].value_counts(),autopct = '%1.1f%%',labels=labels,explode=[0.05,0.05,0.05,0.05])
plt.title('Payment Method')
plt.show()
```

Electronic check is the most used payment method followed by mailed check

```python
plt.figure(figsize=(15,8))
plt.subplot(121)
sns.countplot(x=df['PaymentMethod'],hue=df['Contract']);
sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})
plt.title('Contract based payment method ')
plt.xticks(rotation=60)
plt.show()
plt.figure(figsize=(15,8))
plt.subplot(122)
sns.countplot(x=df['PaymentMethod'],hue=df['gender']);
sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})
plt.title('Gender based payment method ')
plt.xticks(rotation=60)
plt.show()
```

contract based and gender based payment details are shown above
